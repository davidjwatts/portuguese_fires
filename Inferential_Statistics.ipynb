```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import utils
%matplotlib inline
df = utils.wrangle()
```

```python

plt.subplots(figsize=(16,9))
correlation_mat = df[df.columns[4:]].corr()
_ = sns.heatmap(correlation_mat, annot=True)

```

Correlation coefficients are quite low all around. The highest any two predictors get in correlation is 0.68, which is still quite low in terms of colinearity. The highest correlation between the predictors and either area or log(area) is 0.098, which extremely low.
